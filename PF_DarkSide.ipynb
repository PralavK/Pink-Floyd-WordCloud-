```python
import requests
from bs4 import BeautifulSoup
```

```python
source12 = requests.get('http://www.myths.com/pub/lyrics/Pink-Floyd_1_dsotm.html')
soup12 = BeautifulSoup(source12.text, 'html.parser')
```

```python
lyrics = soup12.find_all("pre")
```

```python
dark =""
for lines in lyrics:
    dark = dark+lines.text
    
dark 
```

```python
dark = dark.replace("\n", " ")
```

```python
# There were alot of songs which only had instrumental, so decided to remove it
dark = dark.replace("[Instrumental]", " ").lower()
```

```python
from wordcloud import WordCloud, STOPWORDS
```

```python
stopwords = set(STOPWORDS)
len(stopwords)
Dark = WordCloud(background_color="black", stopwords=stopwords)
Dark.generate(dark)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(10) # set height


plt.imshow(Dark, interpolation='bilinear')
plt.axis('off')
plt.show()
```
