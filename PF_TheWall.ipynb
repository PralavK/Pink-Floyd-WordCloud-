```python
import requests
from bs4 import BeautifulSoup
```

```python
source11 = requests.get('http://www.myths.com/pub/lyrics/Pink-Floyd_3_tw.html')
soup11 = BeautifulSoup(source11.text, 'html.parser')
```

```python
lyr =""
wall = soup11.find_all("pre")
for x in wall:
    lyr = lyr+x.text
lyr = lyr.replace("\n", " ").lower()
```

```python
from wordcloud import WordCloud, STOPWORDS
```

```python
stopwords = set(STOPWORDS)
```

```python
Wall = WordCloud(background_color="black", stopwords=stopwords)
```

```python
Wall.generate(lyr)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(12) # set height


plt.imshow(Wall, interpolation='bilinear')
plt.axis('off')
plt.show()
```
