```python
import requests
from bs4 import BeautifulSoup
```

```python
srcs = requests.get("http://www.myths.com/pub/lyrics/Pink-Floyd_2_wywh.html")
soup12 = BeautifulSoup(srcs.text, "html.parser")
```

```python
#Appending lyrics as strings 
wywh = ""
lyrcs =  soup12.find_all("pre")
for x in lyrcs:
    wywh= wywh+(x.text)
wywh = wywh.replace("\n", " ")
```

```python
#Lowering all words to a smaller case
wywh.lower()
```

```python
from wordcloud import WordCloud, STOPWORDS
```

```python
stopwords = set(STOPWORDS)
```

```python
Shine = WordCloud(background_color='black', stopwords=stopwords)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Shine.generate(wywh)
```

```python
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(10) # set height

# display the cloud
plt.imshow(Shine, interpolation='bilinear')
plt.axis('off')
plt.show()
```
